```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
customers = pd.read_csv("Customers.csv")
products = pd.read_csv("Products.csv")
transactions = pd.read_csv("Transactions.csv")
```

```python
# Display the first few rows of each dataset
print("Customers Dataset:")
print(customers.head(), "\n")
```

```python
print("Products Dataset:")
print(products.head(), "\n")
```

```python
print("Transactions Dataset:")
print(transactions.head(), "\n")
```

```python
# Overview of datasets
print("Customers Info:")
print(customers.info(), "\n")
```

```python
print("Products Info:")
print(products.info(), "\n")
```

```python
print("Transactions Info:")
print(transactions.info(), "\n")
```

```python
# Check for missing values
print("Missing Values in Customers Dataset:")
print(customers.isnull().sum(), "\n")
```

```python
print("Missing Values in Products Dataset:")
print(products.isnull().sum(), "\n")
```

```python
print("Missing Values in Transactions Dataset:")
print(transactions.isnull().sum(), "\n")
```

```python
# Descriptive statistics of transactions dataset
print("Transactions Summary Statistics:")
print(transactions.describe(), "\n")
```

```python
# Handle missing values (if any) - Example: Forward-fill missing data
customers.ffill(inplace=True)
products.ffill(inplace=True)
transactions.ffill(inplace=True)
```

```python
# Merge the datasets
merged_data = transactions.merge(customers, on="CustomerID").merge(products, on="ProductID")
```

```python
# Check merged dataset info and sample rows
print("Merged Dataset Info:")
print(merged_data.info(), "\n")
```

```python
print("Merged Dataset Sample Rows:")
print(merged_data.head(), "\n")
```

```python
# Save merged data to a CSV file
merged_data.to_csv("merged_data.csv", index=False)
```

```python
# Key metrics
total_transactions = len(merged_data)
unique_customers = merged_data["CustomerID"].nunique()
unique_products = merged_data["ProductID"].nunique()
```

```python
print(f"Total Transactions: {total_transactions}")
print(f"Unique Customers: {unique_customers}")
print(f"Unique Products: {unique_products}\n")
```

```python
# Region-wise sales
region_sales = merged_data.groupby("Region")["TotalValue"].sum().sort_values(ascending=False)
print("Region-wise Total Sales:")
print(region_sales, "\n")
```

```python
# Visualization: Region-wise sales
region_sales.plot(kind="bar", figsize=(10, 5), color="skyblue")
plt.title("Region-wise Sales")
plt.xlabel("Region")
plt.ylabel("Total Sales Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Category-wise total sales
category_sales = merged_data.groupby("Category")["TotalValue"].sum()
print("Category-wise Total Sales:")
print(category_sales, "\n")
```

```python
# Visualization: Category-wise sales
category_sales.plot(kind="bar", figsize=(10, 5), color="orange")
plt.title("Category Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 products by quantity sold
top_products = merged_data.groupby("ProductName")["Quantity"].sum().sort_values(ascending=False).head(10)
print("Top 10 Products by Quantity Sold:")
print(top_products, "\n")
```

```python
# Visualization: Top 10 products
top_products.plot(kind="bar", figsize=(10, 5), color="green")
plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Product Name")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Check for duplicate rows in datasets
print("Duplicate Records:")
print(f"Customers: {customers.duplicated().sum()} duplicates")
print(f"Products: {products.duplicated().sum()} duplicates")
print(f"Transactions: {transactions.duplicated().sum()} duplicates\n")
```

```python
# Filter numeric columns only
numeric_data = merged_data.select_dtypes(include=["number"])

# Check if numeric data exists
if not numeric_data.empty:
    plt.figure(figsize=(10, 6))
    correlation = numeric_data.corr()
    sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Correlation Heatmap")
    plt.tight_layout()
    plt.show()
else:
    print("No numeric columns available for correlation heatmap.")
```

```python
# Sales Trends (if date data exists)
if "TransactionDate" in transactions.columns:
    transactions["TransactionDate"] = pd.to_datetime(transactions["TransactionDate"])
    sales_trend = transactions.groupby(transactions["TransactionDate"].dt.to_period("M"))["TotalValue"].sum()
    sales_trend.plot(kind="line", figsize=(10, 5), marker="o")
    plt.title("Sales Trend Over Time")
    plt.xlabel("Month")
    plt.ylabel("Total Sales Value")
    plt.tight_layout()
    plt.show()
```
